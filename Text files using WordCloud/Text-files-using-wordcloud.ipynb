```python
import numpy as np
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
```

# Creating a word cloud for Batman script

```python
bat = open('batman.txt', 'r', encoding='utf-8').read()
```

```python
batmask = np.array(Image.open('batman.png'))
stopwords= set(STOPWORDS)
```

```python
wc = WordCloud(background_color='white',contour_width = 3,mask= batmask, stopwords= stopwords).generate(bat)
```

```python
fig= plt.figure(figsize=(20,30))
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('batman3.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Alice Novel

```python
novel = open('alicenovel.txt','r', encoding='utf-8').read()
image = np.array(Image.open('alice_mask.png'))
stopwords = set(STOPWORDS)
```

```python
alice = WordCloud(background_color='white', mask=image, stopwords=stopwords).generate(novel)
```

```python
fig=plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice,interpolation='bilinear')
plt.axis('off')
plt.savefig('alice.png', dpi=300, bbox_inches='tight')
plt.show()
```
